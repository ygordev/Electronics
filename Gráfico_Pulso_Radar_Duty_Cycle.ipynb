```python
from scipy import signal
import matplotlib.pyplot as plt
import numpy as np

frequencia_moduladora = 20*(10**3) # 20KHz
FRP = 1*(10**3) # 1kHz
fator_trabalho = 0.5 # 20%

periodo_moduladora = 1/frequencia_moduladora
intervalo_repeticao_pulsos = 1/FRP

sinal = []

# Geração do Pulso
t_pulso = np.linspace(0, 3*(10**-3), 7000, dtype=float)
pulso = signal.square(2* np.pi * FRP * t_pulso, duty=fator_trabalho)
plt.plot(t_pulso, pulso)
plt.ylabel('Amplitude')
plt.xlabel('Tempo (s)')
plt.show()

# Geração da Portadora
t_carrier = np.linspace(0, 3*(10**-3), 7000, dtype=float)
sig = np.sin(2 * np.pi * frequencia_moduladora * t_carrier)
plt.plot(t_carrier, sig, 'r')
plt.ylabel('Amplitude')
plt.xlabel('Tempo (s)')
plt.show()

# Geração do Sinal
for i in range(len(t_carrier)):
  if(pulso[i] == -1.0):
    sinal.append(0)
  else:
    sinal.append(abs(sig[i]))

t_sinal = np.linspace(0, 3*(10**-3), len(sinal), dtype=float)
plt.plot(t_sinal, sinal, 'g')
plt.ylabel('Amplitude')
plt.xlabel('Tempo (s)')
plt.show()

```
